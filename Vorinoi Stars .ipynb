```python
def region_for_point(index, vorin):
    return np.array([ vorin.vertices[a] for a in vorin.regions[vorin.point_region[index]]])
```

```python
def plot_region(region, lines, point):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    plt.scatter(region[:,0], region[:,1], zs=region[:,2],  marker='o', zdir='z')
    plt.scatter(point[0], point[1], zs=point[2], marker='o',c='red',zdir='z')
```

```python
def line_segments(vor, point_no):
    line_segments = []
    region = vor.regions[v.point_region[point_no]]
    for simplex in vor.ridge_vertices:
        if np.all(simplex >=0) :
            for_point = np.intersect1d( simplex, region ) 
            if len(for_point)>0:
                line_segments.append([(x,y,z) for x,y,z in vor.vertices[for_point]])
    return line_segments
    
```

```python
segments = line_segments(v,40)
segments
```

```python
def plot_segments(segments):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    for seg in segments:
        s= np.array(seg)
        plt.plot(s[:,0],s[:,1],zs=s[:,2],zdir='z')
        
```

```python
def extr_base():
    return [ Point3(0.05*cos(x),0.05*sin(x),0) for x in np.linspace(0,3.141*2, 20)]
```

```python
extrude_base = extr_base()
```

```python
def print_structs(segments, filename):
    shapes =[]
    for index, seg  in enumerate(segments):
        shapes.append(extrude_along_path(extrude_base,seg) )
        model = union()(shapes)
        with open(filename,'w') as f:
            f.write(scad_render(model))
```

## Hulls

```python
from scipy.spatial import ConvexHull
from subprocess import call
```

```python
def hull_for_star(v, i):
    return ConvexHull(region_for_point(i,v))
```

```python
def print_hull(i,v,filename_scad,filename_stl):
    points = region_for_point(i,v)
    trans='\n'.join(['translate([{0},{1},{2}])cube(0.1,0.1,0.1);'.format(a[0]*100,a[1]*100,a[2]*100) for a in points])
    with open(filename_scad, 'w') as f: 
        f.write("hull(){{ {0} }}".format(trans))
    call(['/Applications/OpenSCAD.app/Contents/MacOS/OpenSCAD',
          '-o', 
          filename_stl, 
          filename_scad])   
```

```python
for star in range(100):
    print 'doing {0} of {1}'.format(star, len(v.points))
    filename_scad = 'Solids/scad/{star}.scad'.format(star=star)
    filename_stl = 'Solids/stl/{star}.stl'.format(star=star)
    print_hull(star,v,filename_scad,filename_stl)
```
